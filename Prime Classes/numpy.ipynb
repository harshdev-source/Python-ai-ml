```python
import numpy as np
import time
```

```python
# Execution performance

size = 1_000_000

# Will measure time
py_list = list(range(size)) 
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"Python list time = {end-start} seconds")

# Time measure for numPY Array
np_arr = np.array(py_list)
start2 = time.time()
sq_array = np_arr ** 2
end2 = time.time()
print(f"NumPy array list time = {end2-start2}")
```

```python
# Memory
import sys

print(f"python list size = {sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"np array size = {np_arr.nbytes} bytes")
```
