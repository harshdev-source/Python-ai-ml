```python
import pandas as pd
```

```python
info = {
    "name": ["Rahul","harshita","aprajita"],
    "CGPA":[9.8, 8.8, 10]
}

df = pd.DataFrame(info)
print(df)
df
```

```python
# Series
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))

# Index 
print(s[1])
print(s[2])
```

```python
# Labelling in Series

s = pd.Series([20, 18, 26, 40], index = ["Harsh", "Rohit", "Poonam", "Usha"])
# print(s[0])
print(s.index)
```

```python
# Vectorized Operation:-
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

print(s1+s2)

# Mutuable values
s1[0] = 87
print(s1)

changed_s1 = s1.drop(0)
print(changed_s1)
print(s1)
```

```python
# DataFrame
info = {
    "name" : ["Adam", "Eve", "Bob"],
    "age" : [23, 35, 20],
    "GPA" : [9.5, 8.6, 7.2]
}

df = pd.DataFrame(info)
print(df.index)
print(df.columns)
df
```

```python
# Lists of list
df = pd.DataFrame([["Adam",28], ["Bob",20], ["Eve", 40]],columns = ["Name", "Age"])
df
```

```python
# DataFrame from numoy arrays
import numpy as np

np_arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(np_arr, columns = ["A","B","C"])
df
```
